```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
```

# Aproximace na okolí bodu

## Taylorův polynom

```python
from numericke_metody.function_approx import taylor
```

```python
from sympy.functions import sin
```

```python
# symbolická proměnná
x = sy.Symbol('x')
# vstupní funkce
f = sin(x)
```

```python
taylor(f, 0, 4)
```

```python
def taylor_plot(f, x0, n_list, x_lims, y_lims, figsize=(10, 4)):
    x = sy.Symbol('x')
    x1 = np.linspace(x_lims[0], x_lims[1], np.abs(x_lims[0] - x_lims[1]) * 20)
    f_lambda = sy.lambdify(x, f, 'numpy')
    plt.figure(figsize=figsize)
    plt.plot(x1, f_lambda(x1), label='Input Function')
    for n in n_list:
        p = taylor(f, x0, n)
        p_lambda = sy.lambdify(x, p, 'numpy')
        plt.plot(x1, p_lambda(x1), label=f'Taylor Series of Order {n}')
    plt.xlim(x_lims)
    plt.ylim(y_lims)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(False)
    plt.title('Taylor Series Approximation')
    plt.show()
```

```python
taylor_plot(f, 0, [2, 4, 6, 8, 14], [-8, 8], [-2, 2])
```
