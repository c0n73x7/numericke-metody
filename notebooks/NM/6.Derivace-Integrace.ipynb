```python
import matplotlib.pyplot as plt
#from numericke_metody.derint.derivate import forward_difference, backward_difference, central_difference
from numericke_metody.derint.derivate import finite_difference
```

```python
help(finite_difference)
```

```python
# fce kterou chci derivovat, musí být v proměnné x
fce = 'sin(1/x)' 

# pro jaké velikosti kroku budu derivovat
hs = [1e-16,1e-15,1e-14,1e-13,1e-12,1e-11,1e-10,1e-09,
     1e-08,1e-07,1e-06,1e-05,0.0001,0.001,0.01,0.1]

res_central = [finite_difference(fce,1,h, method='cd') for h in hs]
res_forward = [finite_difference(fce,1,h, method='fd') for h in hs]
res_backward = [finite_difference(fce,1,h, method='bd') for h in hs]
```

```python
plt.figure(figsize=(12,8))
plt.plot(hs,[p[1]['err'] for p in res_central], 'o-', label='Centralní dif.')
plt.plot(hs,[p[1]['err'] for p in res_forward], '>-', label='Pravá dif.')
plt.plot(hs,[p[1]['err'] for p in res_backward], '<-', label='Levá dif.')
plt.yscale('log')
plt.xscale('log')    
plt.title('Chyba pro různé diference a různé velikosti kroku $h$')
plt.legend()
plt.show()
```
